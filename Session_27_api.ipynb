```python
import pandas as pd
import requests
```

```python
response = requests.get("https://api.themoviedb.org/3/movie/top_rated?api_key=84a7f5bd232fc8cf82c298a6e89569b6")
```

### Now let's extract the useful information that we need

```python
pd.DataFrame(response.json()['results'])[['id','title','release_date','overview','popularity','vote_average','vote_count','poster_path','original_language','adult']]
```

### As we know that was the data from the first page of the tmdb website, as our responses has total 473 pages we have to do the same thing for 473 times, so we'll run a for look for 473 times, we'll crete an empty dataframe and every time we run the loop we'll appen the data into it.

```python
df = pd.DataFrame()
for i in range(1,474):
    response = requests.get("https://api.themoviedb.org/3/movie/top_rated?api_key=84a7f5bd232fc8cf82c298a6e89569b6&language=en-US&page={}".format(i))
    temp_df = pd.DataFrame(response.json()['results'])[['id','title','release_date','overview','popularity','vote_average','vote_count','poster_path','original_language','adult']]
    df = pd.concat([df,temp_df],ignore_index=True)
```
